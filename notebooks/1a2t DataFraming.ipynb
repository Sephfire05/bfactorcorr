```python
# Importing Libraries
from biopandas.pdb import PandasPdb
import pandas as pd # About to change some options 
import numpy as np
# Prior to modifications 
# pd.options.display.max_rows
# Prints 60 as max rows
# pd.set_option("display.max_rows", None)
# print(pd.options.display.max_rows)
```

```python
# Securely fetching the protein and displaying
ppdb = PandasPdb().fetch_pdb('1A2T')
ppdb.df['ATOM'].head()
df = pd.DataFrame(ppdb.df['ATOM'])
```

```python
df.head()
```

```python
def organizeddf(df):
    df1 = df.copy()
    del_cols = ['blank_1', 'alt_loc', "blank_2", 'insertion'
               ,'blank_3', 'occupancy', 'x_coord', 'y_coord', 'z_coord'
               ,'blank_4', 'segment_id', 'charge', 'line_idx']
    sorted_cols = ['record_name', 'atom_name', 'element_symbol', 'residue_name', 'chain_id'
                  ,'residue_number','b_factor']
# This command successfully drops those listed columns
    df1.drop(labels = del_cols,inplace=True, axis=1)
    df1.set_index('atom_number', inplace=True)
    df1 = df1[sorted_cols]
    return df1
```

```python
df1 = organizeddf(df)
df1.head(20)
```

```python
# Returns a data conditioned DF
dfstd = organizeddf(df)
resbdf = dfstd.iloc[:,5:6]
resbdf['residue_number']
residues = resbdf['residue_number'].fillna(0)
print(residues)
#print(residues)
#print(type(residues))
# residues values as full numpy array 
residx = resbdf['residue_number'].unique()
print(residx)
# print(residues) prints all full values as an numpy array
# rescount = pd.Series(data = residues, )
# residues2 just gives me the length of the array of unique elements
# residues2 = resbdf['residue_number'].nunique()
resdfIdx = pd.Index(residx)
resdfval = pd.Series(residues)
#print(resdfIdx)
resdfcounts = resdfval.value_counts()
print(resdfcounts) # This prints out an unsorted index of unique values with the number of index rows it occupies.
```

```python
# Using this DataFrame to show the b values.
resgroupbydf = dfstd.iloc[:,5:]
resgroupbydf.head(20)
# Creating a groupby object that groups by the different residue_number values as index with those appropriate
# b_factors
group = resgroupbydf['b_factor'].groupby(resgroupbydf['residue_number'])
# use b_factor values groups by residue number
group.groups
# Prints off the groupby object from Pandas and shows index rows.
# This explicitly specifies the rows of the DF to each residue unique value
```

### Applying operation on the group to obtain results for each individual residue number.

```python
group.sum().head(10)
# Hand calculated res 7 and obtained 365.98 from the values.
# This groupby function seems to work appropriately.
print(group.mean().head(10))
# Mean gives appropriate answer as well.
print(type(group.mean()))
# Type is a Pandas Series.
```
